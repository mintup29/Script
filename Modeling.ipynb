```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

# Import Data

```python
df = pd.read_csv('/content/digilib_data_cleaned.csv')
df
```

```python
df.info()
```

```python
df = df[['Judul', 'desc_cleaned']]
df
```

#Vectorizing Data

```python
tfidf = TfidfVectorizer(max_features=5000)
```

```python
vectorized_data = tfidf.fit_transform(df['desc_cleaned'].values)
vectorized_data
```

```python
tfidf.get_feature_names_out()
```

```python
vectorized_dataframe = pd.DataFrame(vectorized_data.toarray(), index=df['desc_cleaned'].index.tolist())
vectorized_dataframe
```

# Model

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorized_array = vectorized_dataframe.to_numpy()
vectorized_array
```

```python
vectorized_array.shape
```

```python
similarity = cosine_similarity(vectorized_array)
similarity
```

```python
def recomendation_system(judul):
    id_of_digilib = df[df['Judul']==judul].index[0]
    distances = similarity[id_of_digilib]
    digilib_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:15]
    for digilib_id in digilib_list:
        print('===================')
        print(df.iloc[digilib_id[0]].Judul)
```

```python
df['Judul'].iloc[7]
```

```python
recomendation_system('PREDIKSI HARGA BITCOIN BERDASARKAN SENTIMEN SOSIAL DAN HISTORICAL PRICE MENGGUNAKAN LONG-SHORT TERM MEMMORY (LSTM)')
```
