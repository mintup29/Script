```python
df = df.dropna().reset_index(drop=True)
df.info()
```

```python
df = df.drop_duplicates(subset=['movie']).reset_index(drop=True)
df
```

```python
df.info()
```

```python
df['tags'] = df['Genre'] + ' ' + df['Storyline'] + ' ' + df['plot']
df.drop(['Genre', 'Storyline', 'plot'], axis=1, inplace=True)
df
```

# Preprocessing

```python
movie_df = df.copy()
```

## Remove Punctuation

```python
movie_df['tags_cleaned'] = movie_df['tags'].str.replace(r'[^\w\s]+', '')
movie_df
```

## Case Folding

```python
movie_df['tags_cleaned'] = movie_df['tags_cleaned'].str.lower()
movie_df
```

## Stopword

```python
!pip install nltk
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop = set(stopwords.words('english'))
print(stop)
```

```python
def rem_en(input_txt):
    words = input_txt.split()
    noise_free_words = [word for word in words if word not in stop] 
    noise_free_text = " ".join(noise_free_words) 
    return noise_free_text

movie_df["tags_cleaned"] = movie_df["tags_cleaned"].apply(lambda s: rem_en(s))
movie_df
```

## Lemmazitation

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatiser = WordNetLemmatizer()
```

```python
backup = movie_df.copy()
```

```python
def lemmatize_words(text):
  words = text.split()
  words = [lemmatiser.lemmatize(word,pos='v') for word in words]
  return ' '.join(words)
movie_df["tags_cleaned"] = movie_df["tags_cleaned"].apply(lemmatize_words)
movie_df
```
